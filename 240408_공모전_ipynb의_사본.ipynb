```python
import pandas as pd
df = pd.read_csv('ml_2.csv', encoding = 'utf-8')

df.tail()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

y = df.iloc[:,0].values
X = df.iloc[:, [3,24]] #소득금액, 공연장수 추출

plt.scatter(X.iloc[:11,0], X.iloc[:11,1], color = 'red', marker = 'o', label = 'largeppl')
plt.scatter(X.iloc[11:, 0], X.iloc[11:,1], color = 'blue', marker = 'x', label = 'smallppl')

plt.xlabel('earnings')
plt.ylabel('cultural spaces')
plt.legend(loc = 'upper left')

plt.show()
```
